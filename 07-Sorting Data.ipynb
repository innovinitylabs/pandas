in order to sor this by last name

```python
dft.sort_values(by='last', ascending=False) #for descending
```

to sort a list of columns

```python
dft.sort_values(by=['last', 'first'], ascending=False) 
```

each column in different order. we can give an list of `True` or `False` values

```python
dft.sort_values(by=['last', 'first'], ascending=[False, True], inplace=True) 
dft
```

to bring it back to how it was sorted by index like before  
we use `sort_index()` methos

```python
dft.sort_index()
```

to sort a single column. we can sort a single series as well

```python
dft['last'].sort_values()
```
