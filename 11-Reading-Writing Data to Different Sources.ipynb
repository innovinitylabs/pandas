#### CSV & TSV

#### JSON

<mark>write to JSON</mark>

we need to use`.to_json()` method

```python
india_df.to_json('data/india_df.json')
```

by default it saves as 'dictionary like' JSON where all the responses are within a 'MainBranch'

to save it as 'list-like'

```python
india_df.to_json('data/india_df_list.json', orient='records', lines=True)
```

`lines=True` make each of the records in new line.  
also the json is now saved like list

---

<mark>Read JSON

we need to use the same arguments (setting) we used to write them when reading them

```python
json = pd.read_json('data/india_df_list.json', orient='records', lines=True)
```

```python
json.head(2)
```

#### SQL

<div class="alert alert-box alert-warning"> ** NEED AN SQL DB FOR THIS SECTION

we need to install sequelalchemy library `sqlalchemy` for this. its popular ORM (object relational mapper) to connect SQL with python 

`pip install sqlalchemy`

for postgres need `pscopg2`

`pip install pscopg2-binary `

```python
from sqlalchemy import create_engine
import psycopg2
```

```python
engine = create_engine('postgresql://dbuser:dbpass@localhost:5432/sample_db')
```

database name 'sample_table'

<mark>writing to table in database

```python
india_df.to_sql('sample_table', engine, if_exists='replace')
```

<mark>reading from table in database

```python
sql_df = pd.read_sql('sample_table', engine, index_col='Respondent')
```

```python
sql_df.head()
```

<mark> to run a specific SQL query

```python
sql_df = pd.read_sql_query('SELECT * FROM sample_table', engine, index_col='Respondent')
```

```python
sql_df.head()
```
